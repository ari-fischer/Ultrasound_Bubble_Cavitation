# Import Packages

```python
import nasaPoly
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt
```

# Try out

The next cell shows all the species available in this code

```python
nasaPoly.listSpecies()
```

```python
# let's use water as an example
water=nasaPoly.Species('H2O(L)')
print(water)
```

The next cell gets the coefficients for the NASA polynomial

```python
water._getCoeffs(298)
```

We can also calculate the enthalpy, entropy and gibbs free energy

```python
print(water.h_0(298)) #J/mol
print(water.s_0(298)) #J/mol-K
print(water.g_0(298)) #J/mol
```

If needed, we can calculate the change as well

```python
delta_H_quad=quad(water.cp_0,298,400)[0] #using numerical integration
print(delta_H_quad)

delta_H_nasa=water.h_0(400)-water.h_0(298) #literally just the change
print(delta_H_nasa)
```

Graphical representation of Cp of water from 298 to 400

```python
T_range=np.linspace(298,400,1000)
Cp_range=[water.cp_0(i) for i in T_range]
plt.plot(T_range,Cp_range)
plt.xlabel('T (K)')
plt.ylabel('Cp (J/mol-K)')
plt.legend(['Water'])
plt.show
```
